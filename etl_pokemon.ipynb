```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import csv


pvp = os.path.join("Downloads", "pogo_meta_data.csv")
gen = os.path.join("Downloads", "pokedex_(Update_05.20).csv")


pvp_results = pd.read_csv(pvp)
gen_results = pd.read_csv(gen)

out = pvp_results.append(gen_results)

gen_results.head(10)
```

```python
keep_col = ['name','generation','type_1','type_2']
new_gen_results = gen_results[keep_col]
new_gen_results.to_csv("newGen.csv", index=False)

new_gen_results.head(10)
```

```python
keep_col = ['Pokemon','Score','Type 1','Type 2','Attack','Defense','Stamina','Stat Product']
new_pvp_results = pvp_results[keep_col]
new_pvp_results.to_csv("newpvp.csv", index=False)

new_pvp_results.head(10)
```

```python
df= pd.read_csv('newGen.csv')
df=df.sort_values(['name'], ascending=[True])
df.head(10)
```

```python
df2= pd.read_csv('newpvp.csv')
df2=df2.sort_values(['Pokemon'], ascending=[True])
df2.head(10)
```

```python
result = pd.concat([df, df2], axis=1, join='inner')
display(result)
```

```python
keep_col = ['name','generation','Type 1','Type 2','Score','Attack','Defense','Stamina','Stat Product']
new_res = result[keep_col]
new_res.to_csv("newres.csv", index=False)

new_res.head(10)
```

```python
sort_score= pd.read_csv('newres.csv')
sort_score=sort_score.sort_values(['Score'], ascending=[False])
sort_score.head(10)
```

```python
plt.bar(sort_score['Type 1']legend = true ,sort_score['Score'])
plt.xlabel('Pokemon Type')
plt.ylabel('PvP Score')
plt.title('PvP Score by Pokemon Type')
plt.show()
```

```python
plt.scatter(sort_score['Stat Product'],sort_score['Score'])
plt.xlabel('Stat Product')
plt.ylabel('PvP Score')
plt.title('PvP Score by Stat Product')


plt.show()
```

```python
plt.plot(sort_score['generation'],sort_score['Score'])
plt.xlabel('generation')
plt.ylabel('PvP Score')
plt.title('PvP Score by generation')


plt.show()
```
