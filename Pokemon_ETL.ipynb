```python
# import dependencies
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
```

# Export PvP CSV into DataFrame

```python
pvp_file = "pvp_data.csv"
pvp_df = pd.read_csv(pvp_file)
pvp_df.head()
```

```python
# check pvp_df column names
pvp_df.columns
```

# Transform PvP DataFrame

```python
# filter out only the necessary column data
pvp_cols = ['Pokemon', 'Score', 'Type 1', 'Type 2', 'Attack', 'Defense', 'Stamina', 'Stat Product']
pvp_df_transformed = pvp_df[pvp_cols].copy()

# check data
pvp_df_transformed.head(10)
```

```python
# export to csv
pvp_df_transformed.to_csv("new_pvp.csv", index = False)
```

# Export Pokedex CSV into DataFrame

```python
pokedex_file = "pokedex.csv"
pokedex_df = pd.read_csv(pokedex_file)
pokedex_df.head()
```

```python
# check pokedex_df column names
pokedex_df.columns
```

# Transform Pokedex DataFrame

```python
pokedex_cols = ['name','generation','type_1','type_2',]
pokedex_df_transformed = pokedex_df[pokedex_cols]
pokedex_df_transformed.head(10)
```

```python
# export to csv
pokedex_df_transformed.to_csv("new_pokedex.csv", index = False)
```

```python
# filter pokedex to only show gen 1 pokemon
#gen1_pokedex = pokedex_df_transformed.loc[pokedex_df_transformed['generation'] == 1]
#gen1_pokedex
```

# Merge the two DataFrames

```python
# left = gen1_pokedex
# right = pvp_df_transformed
merged_df = pokedex_df_transformed.merge(pvp_df_transformed, how = "inner", left_on = 'name', right_on = 'Pokemon')
merged_df.head(10)
```

```python
merged_df
```

```python
merged_df_dropped = merged_df.drop(columns=['Pokemon', 'Type 1', 'Type 2'])
merged_df_dropped
merged_df_dropped.to_csv("merged.csv", index = False)
```

```python
sorted_df = pd.read_csv("merged.csv")
sorted_df = sorted_df.sort_values(['Score'], ascending = False)
sorted_df
```

# Create graphs

```python
# Generate a line plot by generation
plt.scatter(sorted_df['generation'],sorted_df['Score'])
plt.xlabel('Generation')
plt.ylabel('PvP Score')
plt.title('PvP Score by Generation')
plt.show()
```

```python
# create bar plot by generation
plt.bar(sorted_df['generation'],sorted_df['Score'])
plt.xlabel('Generation')
plt.ylabel('PvP Score')
plt.title('PvP Score by Generation')
plt.show()
```

```python
# create plot for score by type
#def cm_to_inch(value):
#    return value/2.54

plt.bar(sorted_df['type_1'],sorted_df['Score'])
plt.xlabel('Pokemon Type')
plt.ylabel('PvP Score')
plt.title('PvP Score by Pokemon Type')
plt.figure(figsize=(8, 6), dpi=400)

# save image
plt.savefig("pvp_type.png")

plt.show()
```
